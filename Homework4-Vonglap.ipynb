#### Bài 8: ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
n = int(input())
def reverse(num):
    rev = 0
    while num > 0:
        rev = rev * 10 + (num % 10) #str(n)[::-1]
        num //= 10
    return rev
def palindrome(num):
    return num == reverse(num)

buoc = 0

while not palindrome(n):
    n = n + reverse(n)
    buoc += 1
print(buoc, n)
```

Bài 10: "[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất."

```python
def tinh_do_dai_collatz(k):
    do_dai = 1
    while k > 1:
        if k % 2 == 0:
            k //= 2
        else:
            k = 3 * k + 1
        do_dai += 1
    return do_dai

n = int(input("Nhập n: "))

max_L = 0
best_x = 0

for i in range(1, n + 1):
    L = tinh_do_dai_collatz(i)
    if L > max_L:
        max_L = L
        best_x = i

print(f"{best_x} {max_L}")
```

Bài 26: [Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 

```python
A = int(input("Nhập số nguyên dương A: "))
f1 = 1
f2 = 1

if A < 1:
    print("Không có số Fibonacci nào thỏa mãn.")
elif A == 1:
    print("Số Fibonacci lớn nhất không vượt quá A là: 1")
else:
    f_lon_nhat = 1
    
    while True:
        f_tiep = f1 + f2
        
        if f_tiep > A:
            f_lon_nhat = f2
            break
        f1 = f2
        f2 = f_tiep

    print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {f_lon_nhat}")
```
