Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
n = int(input("Nhập n: "))
step = 0
while True:
    s = str(n)
    if s == s[::-1]: # Kiểm tra đối xứng 
        break
    n += int(s[::-1])
    step += 1
print(f"{step} {n}")
```

"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất." 

```python
n=int(input("Nhập n : "))
max_len=0
best_x=1
for i in range(1, n+1):
    bientam=i
    dodai=1
    while bientam!=1:
        if bientam%2==0:
            bientam//=2
        else:
            bientam=3*bientam+1
        dodai+=1
    if dodai>max_len:
        max_len=dodai
        best_x=i
print(f"{best_x} {max_len}")
```

```python
n = int(input("Nhập n: "))
count = 0
for i in range(1, int(n**0.5) + 1):
    if n % i == 0:
        if i % 2 == 0: count += 1
        if (n // i) % 2 == 0 and (n // i) != i: count += 1
print(count)
```

```python
a = int(input("Số thứ nhất: "))  # Nhập số a
b = int(input("Số thứ hai: "))   # Nhập số b
def tonguoc(num):
    tong=0
    for i in range(1, num):
        if num%i==0:
            tong+=i
    return tong
if tonguoc(a)==b and tonguoc(b)==a:
    print("true")
else:    print("false")

```

[GCD]
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
tìm ước chung lớn nhất của hai số m và n. 

```python
m = int(input("Nhập m: ")) # Nhập số thứ nhất
n = int(input("Nhập n: ")) # Nhập số thứ hai

# Dùng vòng lặp while để "gọt" số n cho đến khi nó bằng 0
while n != 0:
    so_du = m % n  # Bước 1: Tìm phần dư khi lấy m chia cho n
    m = n          # Bước 2: Gán n cho m (số nhỏ hơn giờ đóng vai trò số lớn)
    n = so_du      # Bước 3: Gán phần dư cho n (để tiếp tục gọt ở vòng lặp sau)

# Khi n bằng 0, thì m chính là Ước chung lớn nhất
print(f"Ước chung lớn nhất là: {m}")
```

Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số
con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và
số lượng chó. Ngược lại, in ra "invalid".

```python
tong_con = int(input("Tổng số con: "))   # Nhập tổng số con 
tong_chan = int(input("Tổng số chân: ")) # Nhập tổng số chân 
found = False                            # Biến đánh dấu đã tìm thấy đáp số chưa

for ga in range(tong_con + 1):           # Giả sử số gà chạy từ 0 đến tổng số con
    cho = tong_con - ga                  # Số chó sẽ là phần còn lại
    # Kiểm tra điều kiện: (Gà * 2 chân) + (Chó * 4 chân)
    if ga * 2 + cho * 4 == tong_chan:
        print(f"Gà: {ga}, Chó: {cho}")   # In ra đáp án nếu khớp số chân
        found = True                     # Đánh dấu là đã tìm thấy
        break                            # Thoát vòng lặp luôn cho nhanh

if not found:                            # Nếu chạy hết vòng lặp mà found vẫn là False
    print("invalid")                     # Thì báo bài toán không có đáp án
```
