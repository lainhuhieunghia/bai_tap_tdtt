Hoán đổi hai số không sd biến tạm thời
Sử dụng phép toán XOR trên bit để hoán đổi giá trị của hai biến.
Đọc và tự tìm hiểu các phép toán trên bit: AND, OR, XOR

```python
# Nhập hai số a và b
a = int(input("Nhập số a: "))
b = int(input("Nhập số b: "))

print(f"Trước khi hoán đổi: a = {a}, b = {b}")

# Sử dụng phép toán XOR để hoán đổi
if a != b: # Tránh trường hợp XOR hai số bằng nhau dẫn đến kết quả bằng 0 (tùy ngôn ngữ)
    a = a ^ b
    b = a ^ b
    a = a ^ b

print(f"Sau khi hoán đổi:  a = {a}, b = {b}")
```

Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2
hay không. 
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0

```python
try:
  # Nhập số nguyên từ bàn phím
  n = int(input("Nhập vào một số nguyên: "))

  # Kiểm tra điều kiện: 
  # 1. n phải lớn hơn 0 (vì lũy thừa của 2 luôn dương)
  # 2. n & (n - 1) phải bằng 0 
  # (điều này có nghĩa là n chỉ có một bit 1 trong biểu diễn nhị phân)  
  if n > 0 and (n & (n - 1)) == 0:
     print(f"{n} là lũy thừa của 2")
  else:
    print(f"{n} không phải là lũy thừa của 2")
except ValueError:
  print("Vui lòng nhập một số nguyên hợp lệ")
```

[Rounding]

nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả
cách nhau một dấu cách



```python

try:

  n = float(input())


  if n == int(n):
    tron_xuong = int(n)
  elif n > 0:
    tron_xuong = int(n)
  else:
    tron_xuong = int(n) - 1


  if n == int(n):
    tron_len = int(n)
  elif n > 0:
    tron_len = int(n) + 1
  else:
    tron_len = int(n)


  if n >= 0:
    gan_nhat = int(n + 0.5)
  else:
    gan_nhat = int(n - 0.5)


  print(f"{tron_len}, {tron_xuong}, {gan_nhat}")
except ValueError:
  print("Vui lòng nhập một số hợp lệ.")
```
