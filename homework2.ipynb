Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diệntíchvàchu vi của nó

```python

try:
  a,b=map(float, input("2 cạnh chữ nhật dài-rộng lần lượt là: ").split())
  if a>=b>=0:
   dien_tich=a*b
   chu_vi=(a+b)*2
   print(f"diện tích bằng: {dien_tich}, chu vi là: {chu_vi}")
  else:
    print("nhập lỗi---->nhập lại")
except ValueError:
   print('nhập lỗi---->nhập lại')
```

```python
def tinh_dt_cv(dai, rong):
    s = dai * rong
    p = (dai + rong) * 2
    return s, p

# Nhập dữ liệu trên cùng một dòng
try:
    d, r = map(float, input("Nhập chiều dài và rộng (cách nhau khoảng trắng): ").split())
    if d > 0 and r > 0:
        dt, cv = tinh_dt_cv(d, r)
        print(f"Kết quả -> Diện tích: {dt}, Chu vi: {cv}")
    else:
        print("Cạnh phải là số dương ")
except ValueError:
    print("nhập sai định dạng số rồi.")
```

```python
class HinhChuNhat:
    def __init__(self, dai, rong):
        self.dai = dai
        self.rong = rong

    def tinh_dien_tich(self):
        return self.dai * self.rong

    def tinh_chu_vi(self):
        return (self.dai + self.rong) * 2
try:
 a=float( input("nhập chiều dài:"))

 b=float( input("nhập chiều rộng :"))
 
 if a >= b >= 0:
  
    hcn = HinhChuNhat(a, b)
    print(f"Diện tích: {hcn.tinh_dien_tich()}")
    print(f"Chu vi: {hcn.tinh_chu_vi()}")

 else:
    print('nhập lại dữ liệu')
except ValueError:
    print('nhập lại dữ liệu')
```

Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
try:
    a, b, c=map(float, input("nhập ba cạnh lần lượt là:").split())
    if a+b>c and a+c>b and b+c>a:
        print('là ba cạnh của một tam giác')
        if a==b==c:
            print('là tam giác đều')
        elif a==b or a==c or b==c:
            print('đây là tam giác cân')
        elif a==(b**2+c**2)**0.5 or b==(a**2+c**2)**0.5 or c==(b**2+a**2)**0.5:
            print('đây là tam giác vuông')
        else:
            print("đây là tam giác thường")
        p=(a+b+c)/2
        dien_tich=(p*(p-a)*(p-b)*(p-c))**0.5
        print(f'diện tích là {dien_tich:.2f}')
    else:
        print("không là ba cạnh của tam giác")
except ValueError:
   print('nhập lại dữ liệu')
        
```

Bài 8: Giải hệ phương trình tuyến tính:
ax + by = m
cx + dy = n

```python

a, b, c, d, m, n = map(float, input("nhập 6 số a,b,c,d, m, n lần lượt là:").split( ))
# Tính các định thức
D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c

# Kiểm tra điều kiện và in kết quả
if D != 0:
    x = Dx / D
    y = Dy / D
    print(f"Hệ có nghiệm duy nhất: x = {x}, y = {y}")
else:
    if Dx == 0 and Dy == 0:
        print("Hệ có vô số nghiệm")
    else:
        print("Hệ vô nghiệm")
```

```python
def rev(n):
    r = 0
    while n > 0:
        r = r * 10 + n % 10
        n //= 10
    return r

n = int(input())
step = 0
while n != rev(n):
    n += rev(n)
    step += 1
print(step, n)
```
