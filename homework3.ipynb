Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 


```python
h = int(input("Nhập độ cao h: "))

for i in range(1, h + 1):
    # 1. In khoảng trắng đệm bên trái để tạo độ nghiêng cho tam giác
    print(" " * (h - i), end="")
    
    # 2. Xử lý phần thân tam giác (vùng chứa dấu sao và lỗ hổng)
    for j in range(1, 2 * i):
        # Điều kiện in dấu sao:
        # j == 1: cạnh trái
        # j == 2*i - 1: cạnh phải
        # i == h: cạnh đáy
        if j == 1 or j == 2 * i - 1 or i == h:
            print("*", end="")
        else:
            print(" ", end="") # In khoảng trống ở giữa
            
    # Xuống dòng sau mỗi tầng
    print()
```

Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k)

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
# Khởi tạo giá trị ban đầu bằng None để đánh dấu chưa có dữ liệu
max_val = None
min_val = None

print("Nhập dãy số nguyên (nhập -1 để dừng):")

while True:
    # Nhập số từ người dùng 
    n = int(input())
    
    # Kiểm tra điều kiện dừng 
    if n == -1:
        break
    
    # Nếu là số hợp lệ đầu tiên, khởi tạo max và min
    if max_val is None:
        max_val = n
        min_val = n
    else:
        # Cập nhật số lớn nhất nếu số mới lớn hơn 
        if n > max_val:
            max_val = n
        # Cập nhật số nhỏ nhất nếu số mới nhỏ hơn 
        if n < min_val:
            min_val = n

# Kiểm tra xem người dùng có nhập số nào hợp lệ trước khi thoát không
if max_val is not None:
    print(f"Số lớn nhất: {max_val}")
    print(f"Số nhỏ nhất: {min_val}")
else:
    print("Bạn chưa nhập số nào ngoài số -1.")
```

Bài 15: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều
cách nhất có thể, ko dùng mảng.

```python
n = int(input("Nhập n: "))
count = 0
temp = abs(n) # Dùng trị tuyệt đối để xử lý cả số âm

if temp == 0:
    count = 1
else:
    while temp > 0:
        temp //= 10  # Chia lấy phần nguyên để loại bỏ chữ số cuối cùng
        count += 1

print(f"Cách 1 - Số chữ số là: {count}")
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu
chữ số lẻ

```python
n = int(input("Nhập số nguyên dương n: "))

# Khởi tạo hai biến đếm
chan = 0
le = 0

# Xử lý trường hợp đặc biệt nếu n = 0
if n == 0:
    chan = 1
else:
    temp = abs(n) # Đảm bảo số dương để vòng lặp chạy đúng
    while temp > 0:
        chu_so = temp % 10  # Lấy chữ số cuối cùng
        
        if chu_so % 2 == 0:
            chan += 1       # Tăng đếm chẵn
        else:
            le += 1         # Tăng đếm lẻ
            
        temp //= 10         # Bỏ chữ số cuối, chuyển sang chữ số tiếp theo

print(f"Số chữ số chẵn: {chan}")
print(f"Số chữ số lẻ: {le}")
```

Bài 17: Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input("Nhập số nguyên dương n: "))
tong = 0
temp = abs(n) # Đề phòng sếp nhập số âm

while temp > 0:
    chu_so = temp % 10  # Lấy chữ số cuối
    tong += chu_so      # Cộng dồn vào biến tổng
    temp //= 10         # Vứt chữ số cuối đi

print(f"Tổng các chữ số của n là: {tong}")
```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay
không

```python
n = int(input("Nhập số nguyên dương n: "))

if n <= 0:
    print(f"{n} không phải là số dạng 3^k")
else:
    temp = n
    # Chừng nào temp còn chia hết cho 3 thì còn chia tiếp
    while temp > 1 and temp % 3 == 0:
        temp //= 3
    
    # Nếu cuối cùng temp bằng 1 thì là dạng 3^k
    if temp == 1:
        print(f"{n} là số dạng 3^k")
    else:
        print(f"{n} không phải là số dạng 3^k")
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input("Nhập số nguyên dương A: "))

if A < 1:
    print("Không có số Fibonacci nào thỏa mãn.")
else:
    f1 = 1
    f2 = 1
    
    # Vòng lặp tìm số Fibonacci tiếp theo
    # Nếu tổng f1 + f2 vẫn <= A thì ta vẫn có thể tiến tiếp trong dãy
    while f1 + f2 <= A:
        f_next = f1 + f2
        f1 = f2
        f2 = f_next
        
    print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {f2}")
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố. 

```python
n = int(input("Nhập số nguyên dương n: "))
count = 0
temp = abs(n) # Xử lý trị tuyệt đối để an toàn

# Nếu n = 0 thì không có chữ số nguyên tố nào (vì 0 không là SNT)
if temp == 0:
    count = 0
else:
    while temp > 0:
        chu_so = temp % 10 # Lấy chữ số cuối cùng
        
        # Kiểm tra xem chữ số đó có phải là 2, 3, 5, hoặc 7 không
        if chu_so == 2 or chu_so == 3 or chu_so == 5 or chu_so == 7:
            count += 1
            
        temp //= 10 # Bỏ chữ số cuối để xét tiếp chữ số bên trái

print(f"Số lượng chữ số là số nguyên tố trong n là: {count}")
```
