Bai 1: Yêu cầu. Cước phí Taxi Thông Minh.
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả,
kiểu số nguyên, đơn vị VND).
Các quy tắc cần cài đặt:
1. Giá mở cửa (0–0.8 km):
- Xe 4 chỗ: 11,000 VND
- Xe 7 chỗ: 12,000 VND
2. Giá theo km sau 0.8 km:
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND.
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường +
chờ).
5. Điều kiện đặc biệt:
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu). 


```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    # 1. Tiền mở cửa và quãng đường
    if car_type == 4:
        fare = 11000 if km <= 0.8 else 11000 + min(km - 0.8, 19.2) * 13000 + max(0, km - 20) * 10500
    else: # Xe 7 chỗ
        fare = 12000 if km <= 0.8 else 12000 + min(km - 0.8, 19.2) * 14000 + max(0, km - 20) * 12000
    
    # 2. Phí chờ (mỗi block 5p sau 5p đầu)
    wait_fee = max(0, (wait_minutes - 1) // 5) * 2000 if wait_minutes > 5 else 0
    
    total = fare + wait_fee
    
    # 3. Phụ thu ban đêm (+20%)
    is_night = start_hour in [22, 23, 0, 1, 2, 3, 4, 5]
    if is_night:
        total *= 1.2
    
    # 4. Giảm giá km > 30 và không phải đêm
    if km > 30 and not is_night:
        total *= 0.9
        
    # 5. Mức tối thiểu
    if total < 50000 and km > 5:
        total = 50000
        
    return int(total)
```

sách được sắp xếp tăng dần.
Quy tắc:
- Nếu a == 0 và b != 0: trả về nghiệm duy nhất [-c / b].
- Nếu a == 0 và b == 0: trả về [] (không có nghiệm hữu hạn duy nhất nếu c == 0, hoặc vô nghiệm nếu c != 0).
- Với a != 0, dùng biệt thức d = bb - 4ac:
- d < 0 → []
- d == 0 → [ -b / (2a) ]
- d > 0 → hai nghiệm [r1, r2] với r1 <= r2

```python
def solve_quadratic(a, b, c):
    if a == 0:
        return [-c/b] if b != 0 else []
    d = b*b - 4*a*c
    if d < 0: return []
    if d == 0: return [-b/(2*a)]
    roots = sorted([(-b + d**0.5)/(2*a), (-b - d**0.5)/(2*a)])
    return roots
```

```python
def C(n, k):
    if k < 0 or k > n:
        return 0
    
    k = min(k, n - k)  # tối ưu
    res = 1
    
    for i in range(1, k + 1):
        res = res * (n - i + 1) // i
    
    return res
```

16 Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3). 

```python
import math
def angle_abc(x1, y1, x2, y2, x3, y3):
    # Vector BA và BC
    ba = (x1-x2, y1-y2)
    bc = (x3-x2, y3-y2)
    dot = ba[0]*bc[0] + ba[1]*bc[1]
    mag_ba = math.sqrt(ba[0]**2 + ba[1]**2)
    mag_bc = math.sqrt(bc[0]**2 + bc[1]**2)
    return math.degrees(math.acos(dot / (mag_ba * mag_bc)))
```

25 Yêu cầu: Viết hàm point_in_quadrilateral(quad, p) trả về True nếu điểm p nằm bên trong hoặc trên biên của tứ giác quad,
và False nếu ngược lại.
- Đầu vào:
- quad: danh sách hoặc bộ gồm đúng 4 đỉnh [(x1,y1), (x2,y2), (x3,y3), (x4,y4)] theo thứ tự (tứ giác có thể lồi hoặc lõm,
nhưng không tự cắt).
- p: một điểm (x, y).
- Đầu ra: giá trị boolean.
- Gợi ý: Cài đặt phép kiểm tra điểm trong đa giác tổng quát (ví dụ: ray casting), đồng thời tính cả trường hợp điểm nằm
trên biên là ở trong. 


```python
def point_in_quadrilateral(quad, p):
    # quad = [(x1,y1), (x2,y2), (x3,y3), (x4,y4)], p = (x,y)
    n = len(quad)
    inside = False
    px, py = p
    for i in range(n):
        x1, y1 = quad[i]
        x2, y2 = quad[(i + 1) % n]
        # Kiểm tra nằm trên biên
        if (min(y1, y2) <= py <= max(y1, y2)) and \
           (x1 + (py - y1) * (x2 - x1) / (y2 - y1) if y2 != y1 else x1) == px:
            return True
        if min(y1, y2) < py <= max(y1, y2):
            x_inters = x1 + (py - y1) * (x2 - x1) / (y2 - y1)
            if px < x_inters:
                inside = not inside
    return inside
```

```python
def on_segment(a, b, p):
    # kiểm tra p có nằm trên đoạn ab không
    (x1, y1), (x2, y2), (x, y) = a, b, p
    
    # kiểm tra thẳng hàng
    if (x - x1)*(y2 - y1) != (y - y1)*(x2 - x1):
        return False
    
    # kiểm tra nằm trong đoạn
    return min(x1, x2) <= x <= max(x1, x2) and min(y1, y2) <= y <= max(y1, y2)


def point_in_quadrilateral(quad, p):
    x, y = p
    n = 4
    inside = False
    
    for i in range(n):
        a = quad[i]
        b = quad[(i + 1) % n]
        
        # nếu nằm trên cạnh → True
        if on_segment(a, b, p):
            return True
        
        x1, y1 = a
        x2, y2 = b
        
        # kiểm tra tia cắt cạnh
        if ((y1 > y) != (y2 > y)):
            x_intersect = x1 + (y - y1) * (x2 - x1) / (y2 - y1)
            if x_intersect > x:
                inside = not inside
    
    return inside
```

26 Yêu cầu: Viết hàm are_collinear(x1, y1, x2, y2, x3, y3) trả về True nếu ba điểm thẳng hàng, ngược lại trả về False.

27 Yêu cầu: Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình ax^2 + bx + c = 0.
- Đầu vào: các số thực a, b, c.
- Đầu ra: 0, 1, hoặc 2. Nếu phương trình suy biến thành 0 = 0 (tức a == 0, b == 0, c == 0), trả về float('inf').
- Quy tắc:
- Nếu a != 0, dùng biệt thức D = b^2 - 4ac.
- Nếu a == 0 và b != 0, đó là phương trình bậc nhất với đúng một nghiệm thực.
- Nếu a == 0 và b == 0 và c != 0, phương trình vô nghiệm. 

```python
def quadratic_roots_count(a, b, c):
    if a == 0:
        if b == 0:
            return float('inf') if c == 0 else 0
        return 1
    d = b**2 - 4*a*c
    if d > 0: return 2
    if d == 0: return 1
    return 0
```

30 Yêu cầu: Viết hàm detect_number_type(s) trả về 'int' nếu s biểu diễn một số nguyên, 'float' nếu biểu diễn một số thực dấu
chấm động, ngược lại trả về 'neither'.
- Đầu vào: chuỗi s.
- Đầu ra: một trong các giá trị {'int', 'float', 'neither'}.
- Quy tắc:
- Cho phép có dấu + hoặc - ở đầu.
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100').
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14').
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này. 

```python
def detect_number_type(s):
    if s.startswith(('+', '-')):
        core = s[1:]
    else: core = s
    
    if core.isdigit(): return 'int'
    
    parts = core.split('.')
    if len(parts) == 2 and parts[0].isdigit() and parts[1].isdigit():
        return 'float'
    
    return 'neither'
```

32 Yêu cầu: Viết hàm is_right_triangle(a, b, c) trả về True nếu a, b, c có thể là ba cạnh của một tam giác vuông, ngược lại trả

```python
def is_right_triangle(a, b, c):
    if not (a+b>c and a+c>b and b+c>a): return False
    s = sorted([a, b, c])
    return abs(s[0]**2 + s[1]**2 - s[2]**2) < 1e-9
```

33 Yêu cầu: Viết hàm compute(a, b, op) tính a (+|-||/) b tùy theo op.
- Đầu vào: các số a, b và ký tự op thuộc {'+', '-', '', '/'}.
- Đầu ra: kết quả số học.
- Yêu cầu thêm:
- Nếu op là '/' và b == 0, phát sinh ValueError.
- Nếu op không hợp lệ, phát sinh ValueError. 


```python
def compute(a, b, op):
    if op == '+': return a + b
    if op == '-': return a - b
    if op == '*': return a * b
    if op == '/':
        if b == 0: raise ValueError("Chia cho 0")
        return a / b
    raise ValueError("Toán tử không hợp lệ")
```

```python
def divide_with_remainder(dividend, divisor):
    return (dividend // divisor, dividend % divisor)
```

36 Yêu cầu: Viết hàm is_product_divisible_by_10(a, b) trả về True nếu a b chia hết cho 10, ngược lại trả về False.
- Đầu vào: số nguyên a, b (có thể âm hoặc bằng 0).
- Đầu ra: giá trị boolean.
- Gợi ý: Một số chia hết cho 10 nếu chữ số tận cùng là 0 (tức là chia hết cho cả 2 và 5). 

```python
def is_product_divisible_by_10(a, b):
    return (a * b) % 10 == 0
```

41 Viết hàm trả về ký tự đứng sau ký tự x n vị trí trong bảng chữ cái.
Giữ nguyên chữ hoa/chữ thường và quay vòng trong phạm vi A..Z hoặc a..z. 

```python
def shift_char(x, n):
    if 'a' <= x <= 'z':
        return chr((ord(x) - ord('a') + n) % 26 + ord('a'))
    if 'A' <= x <= 'Z':
        return chr((ord(x) - ord('A') + n) % 26 + ord('A'))
    return x
```

42 Viết hàm trả về True nếu a, b, c có thể là độ dài ba cạnh của một tam giác; ngược lại trả về False. 


```python
def is_triangle_sides(a, b, c):
    return a+b>c and a+c>b and b+c>a
```

43 Viết hàm tính diện tích tam giác có ba cạnh a, b, c (dùng công thức Heron). Phát sinh ValueError nếu ba cạnh không thể
tạo thành tam giác.

```python
def heron_area(a, b, c):
    if not is_triangle_sides(a, b, c): raise ValueError("Không tạo thành tam giác")
    p = (a + b + c) / 2
    return (p * (p-a) * (p-b) * (p-c))**0.5
```

44 Viết hàm trả về tích các số nguyên từ 1 đến n (giai thừa). Với n=0, trả về 1. 


```python
def factorial(n):
    res = 1
    for i in range(2, n + 1): res *= i
    return res
```

45 Viết hàm trả về số lượng số nguyên dương k sao cho 1 <= k < n và k > (n - 1) / 2. (Ví dụ: n=9 -> 4)

```python
def count_k(n):
    # 1 <= k < n và k > (n-1)/2
    count = 0
    for k in range(1, n):
        if k > (n-1)/2: count += 1
    return count
```
